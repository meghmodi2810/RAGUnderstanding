### Data Ingestion to vector DB

#### Importing libraries

```python
# Required libraries
import os
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from pathlib import Path
```

#### Data Ingestion and data parsing

```python
# Read all the pdf using directory loader
def process_all_pdfs_in_directory(directory_path):
    document_container_list = []
    pdf_dir = Path(directory_path)
    
    pdf_files = list(pdf_dir.glob('**/*.pdf'))
    print(f"Found {len(pdf_files)} PDF files in the directory.")
    
    for pdf_file in pdf_files:
        print(f"Processing file: {pdf_file}")
        try:
            loader = PyPDFLoader(str(pdf_file))
            doc = loader.load()
            
            for documents in doc:
                documents.metadata['source'] = str(pdf_file)
                documents.metadata['file_name'] = pdf_file.stem
                documents.metadata['file_type'] = 'pdf'
                documents.metadata['file_size'] = os.path.getsize(pdf_file)
                
            document_container_list.extend(doc)
            print(f"Loaded {len(doc)} documents from {pdf_file}")
        except Exception as e:
            print(f"Error processing {pdf_file}: {e}")
        
    print(f"Total documents loaded: {len(document_container_list)}")
    return document_container_list
```

```python
all_pdfs = process_all_pdfs_in_directory('../data/')
```

```python
# Alternatively, we can use documentloader for the same purpose. It will be more efficient and faster than the above method. We can use the following code to load all the pdf files in the directory.
# But it will not add the metadata to the documents. We can add the metadata to the documents after loading them using the above method.
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader
def process_all_pdf(directory_path):
    doc = DirectoryLoader(
        directory_path,
        glob='**/*.pdf',
        loader_cls=PyPDFLoader,
        show_progress=False
    ).load()
    return doc

# Add metadata to the documents
def add_metadata_to_documents(documents, directory_path):
    for document in documents:
        source_path = document.metadata.get('source', '')
        if source_path:
            file_path = Path(source_path)
            document.metadata['file_name'] = file_path.stem
            document.metadata['file_type'] = 'pdf'
            document.metadata['file_size'] = os.path.getsize(file_path)
    return documents
```

```python
pdf_document = process_all_pdf('../data/')
pdf_document = add_metadata_to_documents(pdf_document, '../data/')
pdf_document
```

```python
# Read all word files using directory loader
from annotated_types import doc
from langchain_community.document_loaders import Docx2txtLoader
def process_all_word_files(directory_path):
    doc = DirectoryLoader(
        directory_path,
        glob='**/*.docx',
        loader_cls=Docx2txtLoader,
        show_progress=False
    )
    document = doc.load()
    return document

word_document = process_all_word_files('../data/')
word_document
```

#### Chunking

```python
# Chunking the documents into smaller chunks using RecursiveCharacterTextSplitter
def split_documents_into_chunks(documents, chunk_size=1000, chunk_overlap=200):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        length_function=len,
        separators=["\n\n", "\n", " ", ""]
    )
    
    split_docs = text_splitter.split_documents(documents)
    print(f"Total chunks created: {len(split_docs)}")
    
    if split_docs:
        print(f"Example chunk: {split_docs[0].page_content[:500]}...")  # Print the first 500 characters of the first chunk
        
    return split_docs
```

```python
chunked_documents_pdf = split_documents_into_chunks(pdf_document)
chunked_documents_pdf

chunked_documents_word = split_documents_into_chunks(word_document)
chunked_documents_word
```

### Embedding and vectorStore

#### Convertion of text to embeddings

```python
# Importing libraries for embedding and vector store
import numpy as np
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.config import Settings
import uuid
import os
from typing import List, Dict, Any, Tuple
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Embedding model class to handle embedding generation using SentenceTransformer
class EmbeddingModel:
    def __init__(self, model_name: str = 'all-MiniLM-L6-v2'):
        self.model_name = model_name
        self.model = None
        self._load_model()
        
    def _load_model(self):
        try:
            self.model = SentenceTransformer(self.model_name)
            print(f"Loaded embedding model: {self.model_name}")
            print(f"Model embedding dimension: {self.model.get_embedding_dimension()}")
            print(f"Model details: {self.model}")
        except Exception as e:
            print(f"Error loading embedding model {self.model_name}: {e}")
            raise
        
    def encode(self, texts: List[str]) -> np.ndarray:
        try:
            print(f"Encoding {len(texts)} texts with model {self.model_name}...")
            embeddings = self.model.encode(texts, show_progress_bar=True)
            print(f"Successfully encoded {len(texts)} texts. Embedding shape: {embeddings.shape}")
            return embeddings
        except Exception as e:
            print(f"Error encoding texts with model {self.model_name}: {e}")
            raise
```

```python
# Initialize the embedding model
embedding_model = EmbeddingModel(model_name='all-MiniLM-L6-v2')
embedding_model
```

#### Storing of embedding into VectorStoreDB

```python
# Vector store class to handle interactions with Chroma vector database
class VectorStore:
    def __init__(self, collection_name: str = 'pdffiles', persist_directory: str = '../data/chroma_db'):
        self.collection_name = collection_name
        self.persist_directory = persist_directory
        self.client = None
        self.collection = None
        self._initialize_vector_store()
        
    def _initialize_vector_store(self):
        try:
            # Ensure the persistence directory exists and initialize the Chroma client
            os.makedirs(self.persist_directory, exist_ok=True)
            self.client = chromadb.PersistentClient(path=self.persist_directory)
            
            # Create or get the collection for storing PDF document chunks and their embeddings
            self.collection = self.client.get_or_create_collection(
                name=self.collection_name,
                metadata={
                    "description": "Collection for storing PDF document chunks and their embeddings",
                    "hnsw:space": "cosine"
                }
            )
            print(f"Initialized Chroma vector store with collection name: {self.collection_name}")
            print(f"Collection distance metric: {(self.collection.metadata or {}).get('hnsw:space', 'unknown')}")
            
        except Exception as e:
            print(f"Error initializing Chroma vector store: {e}")
            raise
        
    def add_documents(self, documents: List[Dict[str, Any]], embeddings: np.ndarray):
        try:
            if len(documents) != len(embeddings):
                raise ValueError("The number of documents and embeddings must be the same.")
            
            print(f"Adding {len(documents)} documents to the vector store...")
            
            # Prepare data for insertion into the vector store
            ids = []
            metadatas_list = []
            documents_text = []
            embedding_list = []
            
            # Add documents, embeddings, and metadata to the Chroma collection
            for i, (doc, embedding) in enumerate(zip(documents, embeddings)):
                doc_id = str(uuid.uuid4())  # Generate a unique ID for each document
                ids.append(doc_id)
                
                # Add metadata for the document
                metadatas = dict(doc.metadata)  # Convert metadata to a regular dictionary
                metadatas['doc_index'] = i  # Add document index to metadata
                metadatas['embedding_model'] = embedding_model.model_name  # Add embedding model name to metadata
                metadatas['content_length'] = len(doc.page_content)  # Add content length to metadata
                metadatas['chunk_size'] = len(doc.page_content)  # Add chunk size to metadata
                metadatas['source'] = doc.metadata.get('source', 'unknown')  # Add source file path to metadata
                metadatas_list.append(metadatas)  # Append metadata to the list
                
                # document content
                documents_text.append(doc.page_content)
                
                # embedding vector
                embedding_list.append(embedding)
                
            # Add the document, embedding, and metadata to the Chroma collection
            self.collection.add(
                ids=ids,
                embeddings=embedding_list,
                metadatas=metadatas_list,
                documents=documents_text
            )
            print(f"Successfully added {len(documents)} documents to the vector store.")
            print(f"Collection now contains {self.collection.count()} documents.")
            
        except Exception as e:
            print(f"Error adding documents to vector store: {e}")
            raise
```

```python
vector_store = VectorStore(collection_name='pdffiles_cosine', persist_directory='../data/chroma_db')
vector_store
```

```python
# Convert the chunked documents of pdf into a list of dictionaries for adding to the vector store
text = [doc.page_content for doc in chunked_documents_pdf]

embeddings_pdf = embedding_model.encode(text)
vector_store.add_documents(chunked_documents_pdf, embeddings_pdf)

# Convert the chunked documents of word into a list of dictionaries for adding to the vector store
text = [doc.page_content for doc in chunked_documents_word]

embeddings_word = embedding_model.encode(text)
vector_store.add_documents(chunked_documents_word, embeddings_word)
```

### RAG Retrieval Pipeline

```python
# Retrieval class to handle retrieval of relevant documents based on a query
class RAGRetrieval:
    def __init__(self, vector_store: VectorStore, embedding_model: EmbeddingModel):
        self.vector_store = vector_store
        self.embedding_model = embedding_model
        
    def retrieve(self, query: str, top_k: int = 5, score_threshold: float = 0.0) -> List[Dict[str, Any]]:
        print(f"Retrieving documents for query: '{query}'")
        print(f"Top K: {top_k}, Score threshold: {score_threshold}")
        
        # Generate query embedding
        query_embedding = self.embedding_model.encode([query])[0]
        
        # Search in vector store
        try:
            results = self.vector_store.collection.query(
                query_embeddings=[query_embedding.tolist()],
                n_results=top_k
            )
            
            # Process results
            retrieved_docs = []
            
            if results['documents'] and results['documents'][0]:
                documents = results['documents'][0]
                metadatas = results['metadatas'][0]
                distances = results['distances'][0]
                ids = results['ids'][0]
                
                for i, (doc_id, document, metadata, distance) in enumerate(zip(ids, documents, metadatas, distances)):
                    # Convert distance to similarity score (ChromaDB uses cosine distance)
                    similarity_score = 1 - distance
                    
                    if similarity_score >= score_threshold:
                        retrieved_docs.append({
                            'id': doc_id,
                            'content': document,
                            'metadata': metadata,
                            'similarity_score': similarity_score,
                            'distance': distance,
                            'rank': i + 1
                        })
                
                print(f"Retrieved {len(retrieved_docs)} documents (after filtering)")
            else:
                print("No documents found")
            
            return retrieved_docs
            
        except Exception as e:
            print(f"Error during retrieval: {e}")
            return []
```

```python
rag_retrieval = RAGRetrieval(vector_store=vector_store, embedding_model=embedding_model)
rag_retrieval
```

```python
rag_retrieval.retrieve("What are the events of techotsav?", top_k=5, score_threshold=0.0)
```
